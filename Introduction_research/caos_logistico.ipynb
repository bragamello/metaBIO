# Determinismo e caos

Você diria que a equação abaixo é determinística?

$$x_{n+1} = r x_n(1-x_n)$$

**Determinismo:** Em termos simplificados significa que, com essa equação, se você sabe o número inicial $x_0$ e o parâmetro $r$, você sabe em que valor seu sistema irá terminar.

Vamos implementar essa regra de geração de números conhecida como equação logística partindo de um número inicial $x_0$ para diferentes valores de $r$. A ideia é observar o comportamento final, após $N$ iterações

```python
# solução de equações diferenciais acopladas
import numpy as np
import matplotlib.pyplot as plt

# equação logística
def logistic_eq(x,r):
  fx = r*x*(1-x)
  return fx

n = 10000
x = 1e-5*np.ones(n)
r = np.linspace(2,4,n)

N = 10000
last = 100

# plot
fig, ax = plt.subplots()

# N iterações
for i in range(N):
  x = logistic_eq(x,r)
  # manter os últimos pontos
  if i >= N-last:
    ax.plot(r,x, ',k',color='blue', alpha=0.1)

plt.grid(True, which="both", ls="-")
plt.xlabel('r')
plt.ylabel(r'$x_N$')
plt.title('Mapa de bifurcação')
fig.tight_layout()
plt.show()

```
