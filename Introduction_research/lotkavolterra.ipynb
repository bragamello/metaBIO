# As equações de Lotka-Volterra

Modelo para relação presa-predador:

$$\frac{dx}{dt} = \alpha x - \beta xy$$

$$\frac{dy}{dt} = \delta xy - \gamma y$$

Onde $x$ é o número de presas e $y$ o número de predadores (ou densidade)

```python
# solução de equações diferenciais acopladas
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# range do tempo dias
tf = 40
step = 0.1
t = np.arange(0, tf, step)

# parametros de interação entre as espécies
# x: presa
# y: predador
alpha = 2/3 # taxa de crescimento do número de presas
beta = 4/3 # predação
delta = 1 # crescimento do número de predadores
gamma = 1 # mortalidade dos predadores

# número inicial de indivíduos
x0 = 1.9
y0 = 1.9

q0 = np.array([x0,y0])

# equação diferencial
def model(q, t):
  x = q[0]
  y = q[1]
  
  return [alpha*x - beta*x*y, gamma*x*y - delta*y]

# solução da equação diferencial
x1 = odeint(model, q0, t)

# plot
fig, ax = plt.subplots()
plt.plot(t,x1[:,0], color='blue', label = 'Presa')
plt.plot(t,x1[:,1], color='red' ,label = 'Predador')
plt.grid(True, which="both", ls="-")
plt.title('Modelo de Lotka-Volterra')
plt.xlabel('t (dias)')
plt.ylabel('N indivíduos')
plt.legend()
fig.tight_layout()
plt.show()
```

```python
# plot
fig, ax = plt.subplots()

for x0 in np.arange(0.8,1.9,0.1):
  # número inicial de indivíduos
  y0 = x0
  q0 = np.array([x0,y0])
  # solução da equação diferencial
  x1 = odeint(model, q0, t)
  plt.plot(x1[:,0],x1[:,1] ,label = 'x0 = '+str(round(x0,1)))
plt.grid(True, which="both", ls="-")
plt.title('Modelo de Lotka-Volterra \n Espaço de fase')
plt.xlabel('presas')
plt.ylabel('predador')
plt.legend(bbox_to_anchor=(1.35, 1), loc='upper right')
fig.tight_layout()
plt.show()
```
