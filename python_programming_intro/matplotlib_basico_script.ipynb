# **Matplotlib:** Gráficos básicos

# 1. Gráfico de uma função

```python
#Ex 1: Função definida na numpy
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

plt.plot(x,y)
# ajustar o gráfico
plt.title("Gráfico do seno")
plt.xlabel(r"$\theta$ (rad)")
plt.ylabel("sin(x)")
plt.show()
```

```python
#Ex 2: Função qualquer
import matplotlib.pyplot as plt
import numpy as np

def func(x):
  return 1/(1+x**2)

x = np.linspace(-10,10,100)
y = func(x)

plt.plot(x,y, color = 'blue', linestyle = 'dashed', linewidth = 2)
# ajustar o gráfico
plt.title("Gráfico da função")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

# 2. Histograma

```python
import matplotlib.pyplot as plt
import numpy as np

mu = 0
sigma = 1
npts = 1000
x = np.random.normal(mu,sigma,npts)

nbins = 50
plt.hist(x, nbins, color = 'red', histtype='bar', linewidth = 2)
# ajustar o gráfico
plt.title("Histograma")
plt.xlabel(r"x")
plt.ylabel("Distribuição")
plt.show()
```

# 3. Box plot ou violin plot

Resumo da distribuição

```python
import matplotlib.pyplot as plt
import numpy as np

mu = 0
sigma = 1
npts = 1000
x = np.random.normal(mu,sigma,npts)

#plt.violinplot(x)
plt.boxplot(x)
# ajustar o gráfico
plt.ylabel(r"x")
plt.show()
```

Veja o exemplo abaixo: Como mudar as cores? **Camada do Artista!**

```python
import matplotlib.pyplot as plt
import numpy as np

mu = 0
npts = 1000
x = [ np.random.normal(mu,sigma,npts) for sigma in range(1,4)] 
labels = ['x1', 'x2', 'x3']


#plt.violinplot(x)
plt.boxplot(x, labels=labels)
# ajustar o gráfico
plt.ylabel(r"x")
plt.show()
```

# 4. Gráfico em barra

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['homem','mulher'] 
y = [0.8,0.5]

colors = ['blue','red']
width = [0.5,0.2]

#plt.violinplot(x)
plt.bar(x, y, color=colors, width = width)
# ajustar o gráfico
plt.ylabel(r"fração de fumantes")
plt.show()
```
